```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the already filtered CSV 
df = pd.read_csv("Filtered_Data.csv")

# Check first rows
print(df.head(5)
plt.figure(figsize=(10,4))
plt.plot(df["time"], df["value_humidity"], color="blue", label="Humidity (%)")
plt.xlabel("Time")
plt.ylabel("Humidity (%)")
plt.title("Humidity (17:3017:40)")
plt.legend()
plt.grid(True)
plt.show()

# Plot temperature
plt.figure(figsize=(10,4))
plt.plot(df["time"], df["value_temperature"], color="red", label="Temperature (C)")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.title("Temperature (17:3017:40)")
plt.legend()
plt.grid(True)
plt.show()

# Combined plot
plt.figure(figsize=(10,5))
plt.plot(df["time"], df["value_humidity"], label="Humidity (%)", color="blue")
plt.plot(df["time"], df["value_temperature"], label="Temperature (C)", color="red")
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Temperature & Humidity (17:3017:40)")
plt.legend()
plt.grid(True)
plt.show()
```
