```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df_x = pd.read_csv("iPhone Thing-Accelerometer_X.csv")
df_y = pd.read_csv("iPhone Thing-Accelerometer_Y.csv")
df_z = pd.read_csv("iPhone Thing-Accelerometer_Z.csv")

# Make the plots
plt.figure(figsize=(14, 10))

# X
plt.subplot(4, 1, 1)
plt.plot(df_x['time'], df_x['value'], color='red', label='X')
plt.title("Accelerometer X")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# Y
plt.subplot(4, 1, 2)
plt.plot(df_y['time'], df_y['value'], color='green', label='Y')
plt.title("Accelerometer Y")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# Z
plt.subplot(4, 1, 3)
plt.plot(df_z['time'], df_z['value'], color='blue', label='Z')
plt.title("Accelerometer Z")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

# Combined
plt.subplot(4, 1, 4)
plt.plot(df_x['time'], df_x['value'], 'r', label='X')
plt.plot(df_y['time'], df_y['value'], 'g', label='Y')
plt.plot(df_z['time'], df_z['value'], 'b', label='Z')
plt.title("Accelerometer X, Y, Z Combined")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()

plt.tight_layout()
plt.show()
```
