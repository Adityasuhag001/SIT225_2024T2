```python
import serial
import time
from datetime import datetime

# Update the port below to match your Arduino's port
# Example: 'COM3' for Windows or '/dev/ttyUSB0' for Linux/Mac
port = '/dev/cu.usbmodem114201'
baud_rate = 9600
filename = 'hcsr04_data.csv'

# Connect to serial
ser = serial.Serial(port, baud_rate)
time.sleep(2)  # Wait for Arduino to start

print("Started logging... Press Ctrl+C to stop.")

with open(filename, "a") as file:
    try:
        while True:
            line = ser.readline().decode('utf-8').strip()
            if line.replace('.', '', 1).isdigit():
                timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
                file.write(f"{timestamp},{line}\n")
                print(f"{timestamp},{line}")
    except KeyboardInterrupt:
        print("Stopped logging.")
    finally:
        ser.close()
```
