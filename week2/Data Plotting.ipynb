```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Loading the CSV file
df = pd.read_csv("hcsr04_data.csv", header=None, names=["timestamp", "distance"])

# Step 2: Converting the timestamp to datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y%m%d%H%M%S', errors='coerce')

# Step 3: Dropping rows with invalid timestamps
df = df.dropna(subset=['timestamp'])

# Step 4: Converting distance to numeric values
df['distance'] = pd.to_numeric(df['distance'], errors='coerce')

# Step 5: Dropping rows with invalid distance values
df = df.dropna(subset=['distance'])

# Step 6: Plotting distance over time
plt.figure(figsize=(12, 6))
plt.plot(df['timestamp'], df['distance'], marker='o', linestyle='-', color='red')
plt.title("HC-SR04 Distance Readings Over Time")
plt.xlabel("Time")
plt.ylabel("Distance (cm)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
