```python
# Activity 1 - Ascending line
# I'm making a simple line that goes up as x increases.
# Basically, y = x. This will give a straight line moving upward.

import matplotlib.pyplot as plt

x_values = list(range(100))  # x from 0 to 99
y_values = [i for i in range(100)]  # y also goes from 0 to 99

plt.plot(x_values, y_values)
plt.title("Ascending Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis (increasing)")
plt.grid(True)
plt.show()
```

```python
# Activity 2 - Descending line
# Now I want the line to go down as x increases.
# So I'm starting y from 100 and reducing it to 1.

x_values = list(range(100))  # x from 0 to 99
y_values = [100 - i for i in range(100)]  # y decreases with each x

plt.plot(x_values, y_values)
plt.title("Descending Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis (decreasing)")
plt.grid(True)
plt.show()
```

```python
# Activity 3 - Wave using sine
# This one is a wave pattern using math.sin().
# I multiplied the sine value to make it taller and added 50 to move it up (so it stays positive).

import math
import matplotlib.pyplot as plt

x_values = list(range(100))
y_values = [math.sin(i * 0.2) * 50 + 50 for i in x_values]  # scale + shift the wave

plt.plot(x_values, y_values)
plt.title("Wave Pattern (Sine Function)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis (sine wave)")
plt.grid(True)
plt.show()
```
